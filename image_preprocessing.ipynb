Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/emma.jpg')
img_resize = img.resize((200,200))
```

```python
img_resize.save('emma_resize.jpg')
```

```python
#display the image from numpy array

img_res = mpimg.imread('/content/emma_resize.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

```python
print(img_res)
```

***Converting RGB image to Grayscale image using OpenCV***

```python
#importing openCV library 
import cv2
```

```python
img = cv2.imread('/content/emma_1.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```
